# Streamlit 

Une étape importante dans le traitement de données est leurs représentations graphiques afin de pouvoir les visualiser et mieux comprendre leurs comportements.

Un outil pratique pour les Data Scientists ou tout ingénieur en Machine Learning ayant pour objectif de présenter le travail fourni aux autres équipes de leur entreprise est la librairie Python nommée Streamlit.

## Qu’est-ce que Streamlit ? 

Streamlit transforme les scripts de données en applications Web partageables en quelques minutes sans écrire de code HTML.
Ce module a été créé en 2019.
La dernière version à ce jour est la version 1.14.0 et date du 27 octobre 2022.


<div class="alert alert-block alert-success">C'est un module encore très récent dont les fonctionnalités sont régulièrement mises à jour (à ce jour tous les mois).
</div>


## L’installation

Vérifier que vous avez un IDE ou un autre éditeur de texte ainsi que python 3.7 - python3.10 et la commande pip installer dans votre machine dans le cas contraire il est nécessaire de les installer pour pouvoir installer Streamlit. 

Si vous êtes sur Linux: installer pip avec la commande suivante:

    $ sudo apt-get install python3-pip

Si vous êtes sur Windows:
    Installez anaconda et créez un environnement. 
    Sélectionnez l’icône « ▶ » à côté de votre environnement.
    Sélectionnez « Ouvrir le terminal ».

L’installation de Streamilt se fait facilement via l’installateur de paquets python pip. 

    $ pip install streamlit

Ce module n'est pas fourni avec la distribution Python de base. Avec la distribution Miniconda, vous pouvez rapidement l'installer avec la commande :

    $ conda install -y streamlit

Nous vous conseillons également d'installer pandas, matplotlib et numpy pour une meilleure expérience. En effet, ces modules sont essentiels dans le traitement des données et leurs représentations.



### Le lancement d'une application Streamlit 

Le code de l'application doit être écrit dans un fichier *.py*.
Pour lancer l'application, il faut taper la commande suivante dans le terminal:

    $ streamlit run application.py

où *application.py* est le fichier contenant le code de l'application. 


## Utilisation pratique 
 Pour utiliser streamlit, il faut l'importer : 

    import streamlit as st

## Exemple d'une application simple

Chaque fonction streamlit va correspondre à un widget, c'est-à-dire à un élément graphique d'entrée ou de sortie (un bouton, un champ de texte, une liste multiple, un graphique, etc...)
    import streamlit as st
    import matplotlib.pyplot as plt

    # Ajout d'un titre : 
    st.title("Ta première application Streamlit")

    # Ajout de texte : 
    st.write("Hello World")

    # Ajout de texte en markdown: 
    st.markdown("Le groupe E est **_incroyablement_ fantastique**.")

    # Exemple d'un bouton : 
    if st.button("CLIQUE"):
        st.write("aie!")
    else:
        st.write("ALLEZ CLIQUE ;)")

    # Exemple de graphique utilisant matplotlib
    fig = plt.figure() 
    plt.plot([1, 2, 3, 4, 5]) 
    # Affiche le graphique :
    st.pyplot(fig)

En exécutant sur le terminal :
    
    $ streamlit run app.py

### Comment utiliser Pycovid (notre application) ?

Une fois l'application lancée, vous retrouverez sur votre gauche une side bar contenant deux listes déroulantes: 

- Une pour choisir les différents pays : France, Italie, Angleterre, Espagne, Suède, Allemagne.
- Un menu déroulant où l'utilateur peut choisir de visualiser le type de courbe parmi : Nombre de nouveaux cas, nombre total de cas, nombre de nouveaux décès et nombre total de décès.

Nous avons également la possibilité de choisir la période de visualisation des différentes courbes avec deux calendriers :
- Un calendrier pour la date de début.
- Un calendrier pour la date de fin.

L'utilisateur peux également changer la couleur de la courbe du graphe visualisé à l'aide d'une palette de couleur.
